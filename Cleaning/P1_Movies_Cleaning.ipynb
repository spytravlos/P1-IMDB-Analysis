### We need to:
### 1) convert date_published into date format


```python
import datetime
```

```python
data['date_published'] = data['date_published'].replace('[a-zA-z]', '',regex = True)
data.head()
```

```python
data['date_published'] = pd.to_datetime(data['date_published'])
data.info()
```

### 2) convert budget, usa_gross_income, worlwide_gross_income, year into numeric format

```python
#worlwide_gross_income
data['worlwide_gross_income'] = data['worlwide_gross_income'].fillna(0) #we replace nan with 0 here as to not lose data for other examinations
```

```python
data['worlwide_gross_income'] = data['worlwide_gross_income'].replace('[^\d\.]', '', regex = True)
```

```python
data['worlwide_gross_income'] = pd.to_numeric(data['worlwide_gross_income'])
```

```python
#usa_gross_income
data['usa_gross_income'] = data['usa_gross_income'].fillna(0)#we replace nan with 0 here as to not lose data fo
data['usa_gross_income'] = data['usa_gross_income'].replace('[^\d\.]', '', regex = True)
data['usa_gross_income'] = pd.to_numeric(data['usa_gross_income'])
```

```python
#budget
data['budget'] = data['budget'].fillna(0)#we replace nan with 0 here as to not lose data fo
data['budget'] = data['budget'].replace('[^\d\.]', '', regex = True)
data['budget'] = pd.to_numeric(data['budget'])
```

```python
data['year'] = data['year'].replace('[^\d\.]', '', regex = True)
data['year'] = pd.to_numeric(data['year'])
```

```python
data.info()
```

```python
data['usa_gross_income'] = data['usa_gross_income'].floordiv(1000000)
data['worlwide_gross_income'] = data['worlwide_gross_income'].floordiv(1000000)
data['budget'] = data['budget'].floordiv(1000000)
```
